```python
import numpy as np
import pandas as pd
```

```python
data = np.random.randint(50,100,size=(5,4))
```

```python
subjects = ['math','math','physics','physics']
terms = ['term1','term2','term1','term2']
classes = ['class1','class1','class2','class2','class2']
students = ['A','B','C','D','E']
```

```python
df = pd.DataFrame(data,\
                  columns = [subjects,terms],\
                 index = [classes,students])
```

```python
df.index.names = ['class','name']
df.columns.names = ['subjects','term']
```

```python
df
```

```python
University = ['CUHK']
subjects = ['math','physics']
terms = ['term1','term2']
tuples_list = [('class1','A'),('class1','B'),('class2','C'),\
               ('class2','D'),('class2','E'),]
```

```python
df = pd.DataFrame(data,\
                 columns = pd.MultiIndex.from_product([University,subjects,terms]),\
                 index = pd.MultiIndex.from_tuples(tuples_list))
```
