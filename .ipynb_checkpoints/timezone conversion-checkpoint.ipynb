```python
from datetime import datetime
from pytz import timezone
import pytz
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('comdty_futures_trading_universe.csv')
df.timezone.unique()
```

```python
def convert_timezone(df,column,timezone):
    return (pd.concat([df_tmp[column].dt.tz_localize(tz).dt.tz_convert(timezone) for tz, df_tmp in df.groupby('timezone')]).astype(str).str[11:19])
```

```python
df['settlement_end_time'] = pd.to_datetime(df['settlement_end_time'])#.astype(str).str[11:19]
df['market_open_time'] = pd.to_datetime(df['market_open_time'])#.astype(str).str[11:19]
df['market_close_time'] = pd.to_datetime(df['market_close_time'])#.astype(str).str[11:19]
```

```python
df['settlement_end_time_UTC'] = convert_timezone(df,'settlement_end_time','UTC')
df['settlement_end_time_SG'] = convert_timezone(df,'settlement_end_time','Asia/Singapore')
df['market_open_time_UTC'] = convert_timezone(df,'market_open_time','UTC')
df['market_open_time_SG'] = convert_timezone(df,'market_open_time','Asia/Singapore')
df['market_close_time_UTC'] = convert_timezone(df,'market_close_time','UTC')
df['market_close_time_SG'] = convert_timezone(df,'market_close_time','Asia/Singapore')
```
