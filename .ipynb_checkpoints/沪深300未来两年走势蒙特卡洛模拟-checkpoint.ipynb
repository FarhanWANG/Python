```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from WindPy import w
w.start()

#从Wind中导出沪深300的交易数据
hs300 = w.wsd("000300.SH","CLOSE","2017-01-01","-1D","")
#转置为DataFrame格式
hs300f = pd.DataFrame(hs300.Data, index=hs300.Fields, columns=hs300.Times).T
#沪深300收益率，即每日涨跌
hs300ret = (hs300f/hs300f.shift(1)-1).dropna()
```

```python
#蒙特卡洛方法

#时间周期，2年，104周
weeks = 104

# Delta
dt = 1/weeks

#mu（drift),即沪深300期望收益的均值
mu = hs300ret.mean()*5

#波动率sigma,即沪深300收益率的标准差
sigma = hs300ret.std()*np.sqrt(5)
```

```python
def stock_monte_carlo(start_price, weeks,mu,sigma):
    #这个公式输入初始价格，周期，mu,sigma,将输出模拟价格的阵列
    
    #定义价格阵列
    price = np.zeros(weeks)
    price[0] = start_price
    #Shock and Drift
    shock = np.zeros(weeks)
    drift = np.zeros(weeks)
    
    # 运行价格阵列
    for x in np.arange(1, weeks):
        
        #计算Shock
        shock[x] = np.random.normal(loc=mu*dt, scale=sigma*np.sqrt(dt))
        #计算Drift
        drift[x] = mu * dt
        #计算价格
        price[x] = price[x-1] + (price[x-1] * (drift[x] + shock[x]))
        
    return price
```

```python
# 获取初始价格
start_price = 5200
plt.figure(figsize=(20,8))
for run in np.arange(10000):
    plt.plot(stock_monte_carlo(start_price,weeks,mu,sigma))
    
plt.xlabel("Weeks")
plt.ylabel("Price")
plt.title("Monte Carlo Analysis for HS300")
```
