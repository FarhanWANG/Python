```python
FILE_PATH = "C:/Users/Farhan/Desktop/3 CSVs SQL Data/"
exchange_list = ["NasdaqGS"]
date_list = dates_11+dates_12+dates_13+dates_14+dates_15+dates_16+dates_17+dates_18+dates_19+dates_20+dates_21
# date_list = ['2020-11-30','2020-12-31','2021-01-31','2021-02-28']
```

```python
data= []
df = pd.DataFrame()
for exchange in exchange_list:
    for date in date_list:
        file_name = FILE_PATH+exchange+"/"+exchange+"_"+date+".csv"
        tmp = pd.read_csv(file_name)
        data.append(tmp)
        df = pd.concat([df,tmp])
df = df.drop(['Unnamed: 0'], axis=1).reset_index()
#df = df.set_index(['tradingitemid','date']).sort_index()
```

# Mannual:
## 1. Decide whether to pre-define value of some raw-data like minority interest. (Too many NaN values for some stocks.
## 2ã€‚ Missing data: IQ_DA_SUPPL_CF (PMAN is computed without DEPI now), IQ_TBV, IQ_TOTAL_CL (No quarterly data -> No GP/Capital Avg)
## 3. Need to fetch daily price to calculate SMA, Price Volatility etc.

```python
df['Change in CA']=df['IQ_TOTAL_CA-0']-df['IQ_TOTAL_CA-4']
df['Change in Cash']=df['IQ_CASH_INVEST-0']-df['IQ_CASH_INVEST-4']
df['Change in CL']=df['IQ_TOTAL_CL-0']-df['IQ_TOTAL_CL-4']
df['Change in LTD']=df['IQ_CURRENT_PORT_DEBT-0']-df['IQ_CURRENT_PORT_DEBT-4']
df['Change in Tax Payable']=df['IQ_INC_TAX_PAY_CURRENT-0']-df['IQ_INC_TAX_PAY_CURRENT-4']
df['STA'] = np.where(df[['Change in CA','Change in Cash','Change in CL','Change in LTD','Change in Tax Payable']].isnull().any(axis=1), 
                     np.nan, 
                     ((df['Change in CA']-df['Change in Cash'])-(df['Change in CL']-df['Change in LTD']-df['Change in Tax Payable'])-df["IQ_DA_CF-0"])/df["IQ_TOTAL_ASSETS-0"])


df['SNOA'] = np.where(df[['IQ_TOTAL_ASSETS-0','IQ_DA_CF-0','IQ_ST_DEBT-0','IQ_CURRENT_PORT_DEBT-0','IQ_LT_DEBT-0','IQ_MINORITY_INTEREST-0','IQ_PREF_EQUITY-0','IQ_TOTAL_COMMON_EQUITY-0']].isnull().any(axis=1), 
                     np.nan, 
                    ((df['IQ_TOTAL_ASSETS-0']-df['IQ_DA_CF-0'])-(df['IQ_TOTAL_ASSETS-0']-df['IQ_ST_DEBT-0']-df['IQ_CURRENT_PORT_DEBT-0']-df['IQ_LT_DEBT-0']\
                                                         -df['IQ_MINORITY_INTEREST-0']-df['IQ_PREF_EQUITY-0']-df['IQ_TOTAL_COMMON_EQUITY-0']))/df['IQ_TOTAL_ASSETS-0'])



df['DSRI']= np.where(df[['IQ_AR-0','IQ_TOTAL_REV-0','IQ_AR-4','IQ_TOTAL_REV-4']].isnull().any(axis=1), 
                     1, 
                    (df['IQ_AR-0']/df['IQ_TOTAL_REV-0'])/(df['IQ_AR-4']/df['IQ_TOTAL_REV-4']))

df['GMI'] = np.where(df[['IQ_GP-0','IQ_TOTAL_REV-0','IQ_GP-4','IQ_TOTAL_REV-4']].isnull().any(axis=1), 
                     1, 
                    (df['IQ_GP-4']/df['IQ_TOTAL_REV-4']) /(df['IQ_GP-0']/df['IQ_TOTAL_REV-0']))
df['AQI']= np.where(df[['IQ_TOTAL_CA-0','IQ_NPPE-0','IQ_LT_INVEST-0','IQ_TOTAL_ASSETS-0','IQ_TOTAL_CA-4','IQ_NPPE-4','IQ_LT_INVEST-4','IQ_TOTAL_ASSETS-4']].isnull().any(axis=1), 
                     1, 
                    (1-((df['IQ_TOTAL_CA-0']+df['IQ_NPPE-0']+df['IQ_LT_INVEST-0'])/df['IQ_TOTAL_ASSETS-0']))/(1-((df['IQ_TOTAL_CA-4']+df['IQ_NPPE-4']+df['IQ_LT_INVEST-4'])/df['IQ_TOTAL_ASSETS-4'])))
df['SGI']= np.where(df[['IQ_TOTAL_REV-0','IQ_TOTAL_REV-4']].isnull().any(axis=1), 
                     1, 
                    df['IQ_TOTAL_REV-0']/df['IQ_TOTAL_REV-4'])
# DEPI
df['SGAI']= np.where(df[['IQ_SGA_SUPPL-0','IQ_TOTAL_REV-0','IQ_SGA_SUPPL-4','IQ_TOTAL_REV-4']].isnull().any(axis=1), 
                     1, 
                    (df['IQ_SGA_SUPPL-0']/df['IQ_TOTAL_REV-0'])/(df['IQ_SGA_SUPPL-4']/df['IQ_TOTAL_REV-4']))
df['Leverage Ratio T']= np.where(df[['IQ_TOTAL_CL-0','IQ_LT_DEBT-0','IQ_TOTAL_ASSETS-0']].isnull().any(axis=1), 
                     np.nan, 
                    (df['IQ_TOTAL_CL-0']+df['IQ_LT_DEBT-0'])/(df['IQ_TOTAL_ASSETS-0']))
df['Leverage Ratio T-4']= np.where(df[['IQ_TOTAL_CL-4','IQ_LT_DEBT-4','IQ_TOTAL_ASSETS-4']].isnull().any(axis=1), 
                     np.nan, 
                    (df['IQ_TOTAL_CL-4']+df['IQ_LT_DEBT-4'])/(df['IQ_TOTAL_ASSETS-4']))
df['LVGI']= np.where(df[['Leverage Ratio T','Leverage Ratio T-4']].isnull().any(axis=1), 
                     1, 
                    (df['Leverage Ratio T'])/(df['Leverage Ratio T-4']))
df['TATA']= np.where(df[['IQ_NI-0','IQ_CASH_OPER-0','IQ_TOTAL_ASSETS-0']].isnull().any(axis=1), 
                     np.nan, 
                    (df['IQ_NI-0'] - df['IQ_CASH_OPER-0'])/df['IQ_TOTAL_ASSETS-0'])
df['PROBM Model'] = -4.84+df['DSRI']*0.92+df['GMI']*0.528+df['AQI']*0.404+ df['SGI']*0.892-df['SGAI']*0.172-df['LVGI']*0.327+df['TATA']*4.679
df['PROBM to PMAN'] = norm.cdf(df['PROBM Model'])

df['Altman Z'] =  df['IQ_Z_SCORE-0'].fillna(df['IQ_Z_SCORE-0_FY'])

#TBV

df['Net Debt/Total Common Equity'] = df['IQ_NET_DEBT-0']/df['IQ_TOTAL_COMMON_EQUITY-0']


df['IAR'] = df['IQ_GW_INTAN-0_FY']/df['IQ_TOTAL_ASSETS-0_FY']


df['ORECTA'] = (df['IQ_TOTAL_ASSETS-0']-df['IQ_CASH_ST_INVEST-0']-df['IQ_AR-0']-df['IQ_INVENTORY-0']-df['IQ_NPPE-0'])/df['IQ_TOTAL_ASSETS-0']

# Avg Rev/GP Growth 


df['GPTA'] = df['IQ_GP-0']/df['IQ_TOTAL_ASSETS-0']
df['GPTA-1'] = df['IQ_GP-1']/df['IQ_TOTAL_ASSETS-1']
df['GPTA-2'] = df['IQ_GP-2']/df['IQ_TOTAL_ASSETS-2']
df['GPTA-3'] = df['IQ_GP-3']/df['IQ_TOTAL_ASSETS-3']
df['GPTA Avg'] = (df['GPTA']+df['GPTA-1']+df['GPTA-2']+df['GPTA-3'])/4

#GP/Capital Avg

df['GPM']=df['IQ_GP-0_QTR']/df['IQ_TOTAL_REV-0_QTR']
df['GPM-1']=df['IQ_GP-1_QTR']/df['IQ_TOTAL_REV-1_QTR']
df['GPM-2']=df['IQ_GP-2_QTR']/df['IQ_TOTAL_REV-2_QTR']
df['GPM-3']=df['IQ_GP-3_QTR']/df['IQ_TOTAL_REV-3_QTR']
df['GPM-4']=df['IQ_GP-4_QTR']/df['IQ_TOTAL_REV-4_QTR']
df['GPM-5']=df['IQ_GP-5_QTR']/df['IQ_TOTAL_REV-5_QTR']
df['GPM-6']=df['IQ_GP-6_QTR']/df['IQ_TOTAL_REV-6_QTR']
df['GPM-7']=df['IQ_GP-7_QTR']/df['IQ_TOTAL_REV-7_QTR']
df['GPM-8']=df['IQ_GP-8_QTR']/df['IQ_TOTAL_REV-8_QTR'] 
df['Avg Margin Growth'] = np.where(df[['GPM','GPM-1','GPM-2','GPM-3','GPM-4','GPM-5','GPM-6','GPM-7','GPM-8']].isnull().any(axis=1), 
                          np.nan, 
                          gmean([df['GPM']/df['GPM-1'],df['GPM-1']/df['GPM-2'],df['GPM-2']/df['GPM-3'],df['GPM-3']/df['GPM-4'],
                                 df['GPM-4']/df['GPM-5'],df['GPM-5']/df['GPM-6'],df['GPM-6']/df['GPM-7'],df['GPM-7']/df['GPM-8']])-1)

df['Margin Stability'] = np.where(df[['GPM','GPM-1','GPM-2','GPM-3','GPM-4','GPM-5','GPM-6','GPM-7','GPM-8']].isnull().any(axis=1), 
                          np.nan, 
                          df[['GPM','GPM-1','GPM-2','GPM-3','GPM-4','GPM-5','GPM-6','GPM-7']].mean(axis = 1)/
                                 df[['GPM','GPM-1','GPM-2','GPM-3','GPM-4','GPM-5','GPM-6','GPM-7']].std(axis = 1))



df['EBITDA M']=df['IQ_EBIT-0_QTR']/df['IQ_TOTAL_REV-0_QTR']
df['EBITDA M-1']=df['IQ_EBIT-1_QTR']/df['IQ_TOTAL_REV-1_QTR']
df['EBITDA M-2']=df['IQ_EBIT-2_QTR']/df['IQ_TOTAL_REV-2_QTR']
df['EBITDA M-3']=df['IQ_EBIT-3_QTR']/df['IQ_TOTAL_REV-3_QTR']
df['EBITDA M-4']=df['IQ_EBIT-4_QTR']/df['IQ_TOTAL_REV-4_QTR']
df['EBITDA M-5']=df['IQ_EBIT-5_QTR']/df['IQ_TOTAL_REV-5_QTR']
df['EBITDA M-6']=df['IQ_EBIT-6_QTR']/df['IQ_TOTAL_REV-6_QTR']
df['EBITDA M-7']=df['IQ_EBIT-7_QTR']/df['IQ_TOTAL_REV-7_QTR']
df['EBITDA M-8']=df['IQ_EBIT-8_QTR']/df['IQ_TOTAL_REV-8_QTR'] 
df['Avg EBITDA Margin Growth'] = np.where(df[['EBITDA M','EBITDA M-1','EBITDA M-2','EBITDA M-3','EBITDA M-4','EBITDA M-5','EBITDA M-6','EBITDA M-7','EBITDA M-8']].isnull().any(axis=1), 
                          np.nan, 
                          gmean([df['EBITDA M']/df['EBITDA M-1'],df['EBITDA M-1']/df['EBITDA M-2'],df['EBITDA M-2']/df['EBITDA M-3'],df['EBITDA M-3']/df['EBITDA M-4'],
                                 df['EBITDA M-4']/df['EBITDA M-5'],df['EBITDA M-5']/df['EBITDA M-6'],df['EBITDA M-6']/df['EBITDA M-7'],df['EBITDA M-7']/df['EBITDA M-8']])-1)
df['EBITDA Margin Stability'] = np.where(df[['EBITDA M','EBITDA M-1','EBITDA M-2','EBITDA M-3','EBITDA M-4','EBITDA M-5','EBITDA M-6','EBITDA M-7','EBITDA M-8']].isnull().any(axis=1), 
                          np.nan, 
                          df[['EBITDA M','EBITDA M-1','EBITDA M-2','EBITDA M-3','EBITDA M-4','EBITDA M-5','EBITDA M-6','EBITDA M-7']].mean(axis = 1)/
                                 df[['EBITDA M','EBITDA M-1','EBITDA M-2','EBITDA M-3','EBITDA M-4','EBITDA M-5','EBITDA M-6','EBITDA M-7']].std(axis = 1))



df['EV/EBITDA'] =df['IQ_TEV']/ df['IQ_EBITDA-0']
df['EV/EBIT'] =df['IQ_TEV']/ df['IQ_EBIT-0']
df['EV'] = df['IQ_TEV']
df['GP'] = df['IQ_GP-0']
df['EV/GP'] = df['EV'] / df['GP']

df
```
