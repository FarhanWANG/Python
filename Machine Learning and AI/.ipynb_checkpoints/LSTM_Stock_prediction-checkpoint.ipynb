# Load the dataset

```python
data=pd.read_csv("TSLA.csv")
print("Number of rows and columns:", data.shape)
data.head(5)
```

```python
train_data = data.iloc[:800, 5:] #select 800 rows and 1 column (open)
test_data = data.iloc[800:, 5:]
```

```python
print(train_data.shape)
train_data.head()
```

```python
print(test_data.shape)
test_data.head()
```

```python
###convert to numpy format
train_data_np=train_data.values
#test_data_np=test_data.values
```

# Normalize the data using Min Max scalar

```python
sc = MinMaxScaler(feature_range = (0, 1)) ##in the range of 0 to 1

tr_scaled = sc.fit_transform(train_data_np)
#tst_scaled = sc.transform(test_data_np)
```

# Reshape the data for LSTM

The input to every LSTM layer must be three-dimensional.

The three dimensions of this input are:

<b>Samples</b>: One sequence is one sample. A batch is comprised of one or more samples.

<b>Time Steps</b>: One time step is one point of observation in the sample.

<b>Features</b>: One feature is one observation at a time step.

For more information check the following link:

https://machinelearningmastery.com/reshape-input-data-long-short-term-memory-networks-keras/

```python
tr_scaled
```

```python
# Creating a data structure with 60 time-steps and 1 output

X_train = []
Y_train = []

for i in range(60, 800):
    X_train.append(tr_scaled[i-60:i, :]) # 0-60, 1-61, 2-62--------
    Y_train.append(tr_scaled[i, 0]) # 60, 61, 62-----
    
X_train, Y_train = np.array(X_train), np.array(Y_train)
print(X_train.shape)
print(Y_train.shape)

#Reshape the data as (#values, #time-steps, #1 dimensional output) ## 800-60 is 740 time steps
# X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], -1))

print("After reshaping of train data:", X_train.shape) # [samples, timesteps, features]
```

```python
X_train.shape[1]
```

# Build the LSTM Model

```python
model = Sequential()

#Adding the first LSTM layer 
model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 2))) #Time series: You must set return_sequences=True when stacking LSTM layers so that the second LSTM layer has a three-dimensional sequence input
#some Dropout regularization
model.add(Dropout(0.2))

# Adding a second LSTM layer 
model.add(LSTM(units = 50, return_sequences = True))
#some Dropout regularization
model.add(Dropout(0.2))

# Adding a third LSTM layer 
model.add(LSTM(units = 50, return_sequences = True))
#some Dropout regularization
model.add(Dropout(0.2))

# Adding a fourth LSTM layer 
model.add(LSTM(units = 50))
#some Dropout regularization
model.add(Dropout(0.2))

# Adding the output layer
model.add(Dense(units = 1))
```

# Compile the model

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error') #use default learning rate = 0.001
model.summary()
```

# Train the LSTM Model

```python
# Fitting the LSTM to the train data
model.fit(X_train, Y_train, epochs = 50, batch_size = 32)
```

```python
# import h5py
# model.save("LSTM.h5") 
```

# Prepare the test data (reshape the data as [samples, timesteps, features])


```python
#train_data = data.iloc[:800, 1:2]
#test_data = data.iloc[800:, 1:2]
data_tot = pd.concat((train_data,test_data), axis = 0)
print(len(data_tot))
print(len(test_data))
#1259- 459-60: 740 ## starting from index 740: to the end which gives 519
input_data = data_tot[len(data_tot) - len(test_data) - 60:].values #convert input data from pandas format to numpy using .values
print(len(input_data))
```

```python
inputs = input_data.copy()
input_scaled = sc.transform(inputs)
```

```python
X_test = []

for i in range(60, 519): #459 observations tst data, 459+60=519
    
    X_test.append(input_scaled[i-60:i,:]) #0-60, 1-61..... 459 times
    
X_test = np.array(X_test)

#Reshape the data as (#values, #time-steps, #1 dimensional output)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 2))

print("After reshaping of test data:", X_test.shape) # [samples, timesteps, features]
  
```

# Make Predictions using the test data

```python
pred_stock_price = model.predict(X_test)
pred_stock_price = sc.inverse_transform(pred_stock_price) ##inverse the min max transform to see actual values of data
```

# Visualize the results 

```python
# Visualising the results using a plot

figure = plt.figure(figsize=(15, 8))

plt.plot(data.loc[800:, "Date"],test_data.values, color = "red", linewidth = 4, label = "Actual Stock Price for Tesla")
plt.plot(data.loc[800:, "Date"],pred_stock_price, color = "green", linewidth = 4, label = "Predicted Stock Price for Tesla")

#([start, ]stop, [step, ])
plt.xticks(np.arange(0,459,50),fontsize = 13) #np.arange--> Return evenly spaced values within a given interval 
plt.yticks(fontsize = 13)

plt.title('Tesla Stock Price Prediction',fontsize = 18)
plt.xlabel('\nTime',fontsize = 18)
plt.ylabel('Tesla Stock Price\n',fontsize = 18)
plt.legend(loc='best',fontsize = 16)
plt.show() #display the plot
```
