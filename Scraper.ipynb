```python
from lxml import html
import requests
  
# Request the page
page = requests.get('https://www.glazziq.com/th/c/l/t-16')
  
# Parsing the page
# (We need to use page.content rather than 
# page.text because html.fromstring implicitly
# expects bytes as input.)
tree = html.fromstring(page.content)  
  
# Get element using XPath
# //*[@id="product-listing"]/div/div[1]/div[2]/a/div/div[1]
product = tree.xpath('//*[@id="product-listing"]/div/div[3]/div[2]/a/div/div[1]/text()')
print(product)
```
